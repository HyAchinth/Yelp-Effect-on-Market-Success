```python
# tokenization with lemmatization

import nltk
import string
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download required NLTK resources
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Get the set of stop words
stop_words = set(stopwords.words('english'))

# Text Preprocessing Function
def cleanup_text(text):
    if pd.isna(text):  # Handle NaN values
        return ""
    
    # Tokenization
    tokenized_text = word_tokenize(text.lower())  # Convert to lowercase and tokenize

    # Remove punctuation
    tokenized_text = [word for word in tokenized_text if word.isalnum()]

    # Remove stopwords
    tokenized_text = [word for word in tokenized_text if word not in stop_words]

    # Lemmatization
    tokenized_text = [lemmatizer.lemmatize(word) for word in tokenized_text]

    return " ".join(tokenized_text)  # Convert back to string format

# Apply cleanup function to the text column
m1["cleaned_text"] = m1["text"].apply(cleanup_text)

# Save the cleaned dataset
m1.to_csv("cleaned_yelp_reviews.csv", index=False)

# Display processed data
print(m1[["text", "cleaned_text"]].head())
```
