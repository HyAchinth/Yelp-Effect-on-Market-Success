```python
from bertopic import BERTopic
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
import numpy as np
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

topic_model = BERTopic(verbose=True)
topics, probs = topic_model.fit_transform(reviews['cleaned_text'])

topic_info = topic_model.get_topic_info()

for topic_id in topic_info['Topic']:
    if topic_id != -1:
        print(f"Topic {topic_id}: {topic_model.get_topic(topic_id)}")

top_10_topics = topic_info[topic_info['Topic'] != -1].sort_values(by='Count', ascending=False).head(10)

for _, row in top_10_topics.iterrows():
    topic_id = row['Topic']
    top_words = topic_model.get_topic(topic_id)
    print(f"Topic {topic_id}: {[word for word, _ in top_words]}")


nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()

reviews['sentiment_score'] = reviews['cleaned_text'].apply(
    lambda x: sid.polarity_scores(x)['compound']
)

def label_sentiment(score):
    if score >= 0.05:
        return 'POSITIVE'
    elif score <= -0.05:
        return 'NEGATIVE'
    else:
        return 'NEUTRAL'

reviews['sentiment_label'] = reviews['sentiment_score'].apply(label_sentiment)

reviews['dominant_topic'] = topics

top_10_topic_ids = top_10_topics['Topic'].tolist()
filtered_reviews = reviews[reviews['dominant_topic'].isin(top_10_topic_ids)]

topic_sentiment = filtered_reviews.groupby('dominant_topic').agg(
    avg_sentiment=('sentiment_score', 'mean'),
    positive_count=('sentiment_label', lambda x: (x == 'POSITIVE').sum()),
    negative_count=('sentiment_label', lambda x: (x == 'NEGATIVE').sum()),
    neutral_count=('sentiment_label', lambda x: (x == 'NEUTRAL').sum()),
    total_docs=('sentiment_label', 'count')
).reset_index()

topic_sentiment = topic_sentiment.merge(
    topic_info[['Topic', 'Name', 'Representation']],
    left_on='dominant_topic',
    right_on='Topic',
    how='left'
)

print("Aggregated sentiment by topic (BERTopic):")
display(topic_sentiment)

print("\nSample rows:")
display(reviews[['dominant_topic', 'sentiment_score', 'sentiment_label']].head())
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

embeddings = topic_model._extract_embeddings(reviews['cleaned_text'])
pca = PCA(n_components=3)
reduced_embeddings = pca.fit_transform(embeddings)
```

```python
import pandas as pd

viz_df = pd.DataFrame({
    'x': reduced_embeddings[:, 0],
    'y': reduced_embeddings[:, 1],
    'z': reduced_embeddings[:, 2],
    'topic': topics,
    'sentiment_score': reviews['sentiment_score'],
    'sentiment_label': reviews['sentiment_label']
})

topic_names = {
    0: "Mexican Food",
    1: "Coffee Shops",
    2: "Pizza",
    3: "Bakery & Desserts",
    4: "Dining & Sides",
    5: "Customer Service",
    6: "Grocery Stores",
    7: "Craft Beer",
    8: "Dining Atmosphere",
    9: "Seafood"
}
viz_df['topic_name'] = viz_df['topic'].map(topic_names)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(viz_df['x'], viz_df['y'], viz_df['z'], c=viz_df['topic'], cmap='viridis', alpha=0.6)

cbar = plt.colorbar(scatter, ax=ax, pad=0.1)
cbar.set_label('Topic')

ax.set_xlabel('PCA Component 1')
ax.set_ylabel('PCA Component 2')
ax.set_zlabel('PCA Component 3')
ax.set_title('3D PCA Visualization of Topics')

for topic_id, topic_name in topic_names.items():
    topic_center = viz_df[viz_df['topic'] == topic_id][['x', 'y', 'z']].mean()
    ax.text(topic_center[0], topic_center[1], topic_center[2], topic_name, fontsize=10, ha='center')

plt.show()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import numpy as np

def recommend_restaurants_by_top_10_topics(query, reviews, topic_model, top_n=5):
    embedder = SentenceTransformer("all-MiniLM-L6-v2")

    topic_info = topic_model.get_topic_info()
    top_10_topics = topic_info[topic_info['Topic'] != -1].sort_values(by='Count', ascending=False).head(10)
    top_10_topic_ids = top_10_topics['Topic'].tolist()

    topic_representations = []
    for topic_id in top_10_topic_ids:
        topic_words = topic_model.get_topic(topic_id)
        if isinstance(topic_words, list):
            topic_representations.append(" ".join([word for word, _ in topic_words]))
        else:
            topic_representations.append("")

    query_embedding = embedder.encode([query], convert_to_tensor=True)
    topic_embeddings = embedder.encode(topic_representations, convert_to_tensor=True)

    similarities = cosine_similarity(query_embedding.cpu().detach().numpy(),
                                     topic_embeddings.cpu().detach().numpy())

    closest_topic_index = np.argmax(similarities[0])
    closest_topic = top_10_topic_ids[closest_topic_index]
    print(f"Query mapped to Topic {closest_topic}: {topic_representations[closest_topic_index]}")

    matched_reviews = reviews[reviews['dominant_topic'] == closest_topic]

    recommended_restaurants = matched_reviews.groupby('name').agg(
        avg_sentiment=('sentiment_score', 'mean'),
        review_count=('name', 'count'),
        major_topic=('dominant_topic', lambda x: x.mode()[0]),
        stars_y=('stars_y', 'mean')
    ).reset_index()

    recommended_restaurants = recommended_restaurants[
        (recommended_restaurants['review_count'] > 2) &
        (recommended_restaurants['avg_sentiment'] > 0.85)
    ]

    recommended_restaurants = recommended_restaurants.sort_values(
        by=['stars_y'], ascending=[False]
    ).head(top_n)

    return recommended_restaurants

query = "I want to get coffee with friends"
recommendations = recommend_restaurants_by_top_10_topics(query, reviews, topic_model, top_n=5)
print("Recommended Restaurants for the Closest Topic:")
display(recommendations[['name','avg_sentiment','stars_y']])
```
